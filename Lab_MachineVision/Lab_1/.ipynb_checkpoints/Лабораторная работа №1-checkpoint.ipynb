```python
#Открытие видео
import cv2
import numpy as np
import time
print( cv2.__version__ )
# создание объекта типа VideoCapture
cap = cv2.VideoCapture('traffic.mp4')
# проверка успешности открытия видеофайла
if (cap.isOpened() == False): 
  print("Ошибка открытия файла")

# получение fps файла
fps = cap.get(cv2.CAP_PROP_FPS)
# print(fps)

while(cap.isOpened()):
  ret, frame = cap.read()

  if ret == True: 
        
    # корректировка скорости вывода
    time.sleep(1/fps)

    # отображение текущего кадра   
    cv2.imshow('frame',frame)  

    # триггер для выхода
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break
 
  # выход из цикла
  else:
    break 
 
# Освобожение памяти
cap.release()
 
# закрытие окна
cv2.destroyAllWindows() 
```

```python
#Нахождение объектов на изображении
import cv2
import numpy as np
import time
print( cv2.__version__ )
# создание объекта типа VideoCapture
cap = cv2.VideoCapture('traffic.mp4')
ret, frame = cap.read()
# проверка успешности открытия видеофайла
if (cap.isOpened() == False): 
  print("Ошибка открытия файла")

# получение fps файла
fps = cap.get(cv2.CAP_PROP_FPS)
# print(fps)


fgbg = cv2.createBackgroundSubtractorMOG2()

while(cap.isOpened()):
  ret, frame = cap.read()

  if ret == True: 
        
    # корректировка скорости вывода
    time.sleep(1/fps)

    # отображение текущего кадра   
    #cv2.imshow('frame',frame)  
    
    # отображение серого видео
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # Перевод массива кадров в черно-белую градацию
    
    fgmask = fgbg.apply(gray)
    cv2.imshow("Objekt",fgmask) # обработанный вариант
    # триггер для выхода
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break
 
  # выход из цикла
  else:
    break 
 
# Освобожение памяти
cap.release()
# закрытие окна
cv2.destroyAllWindows() 
```

```python
#Добавление прямоугольгиков без обработки Canny, Blur и т.д.
import cv2
import numpy as np
import time
print( cv2.__version__ )
# создание объекта типа VideoCapture
cap = cv2.VideoCapture('traffic.mp4')
ret, frame = cap.read()
# проверка успешности открытия видеофайла
if (cap.isOpened() == False): 
  print("Ошибка открытия файла")

# получение fps файла
fps = cap.get(cv2.CAP_PROP_FPS)
# print(fps)


fgbg = cv2.createBackgroundSubtractorMOG2()

while(cap.isOpened()):
  ret, frame = cap.read()

  if ret == True:     
    # корректировка скорости вывода
    time.sleep(1/fps)
   
    # отображение серого видео
    
    grayBlur = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # Перевод массива кадров в черно-белую градацию
      
    fgmask = fgbg.apply(grayBlur)   
    
    contours, hierarchy = cv2.findContours(fgmask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    cv2.line(frame, (390,490), (1200,440), (0,0,255), 5)
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area>1000:
            peri = cv2.arcLength(cnt,True)
            approx = cv2.approxPolyDP(cnt, 0.02*peri, True)
            x, y, w, h = cv2.boundingRect(approx)
            cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 5)
               
    cv2.imshow("Objekt",frame) # обработанный вариант  
        
    # триггер для выхода
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break
 
  # выход из цикла
  else:
    break 
 
# Освобожение памяти
cap.release()
# закрытие окна
cv2.destroyAllWindows() 
```

```python
#Подсчет количества авто         
import cv2
import numpy as np
import time
print( cv2.__version__ )
# создание объекта типа VideoCapture
cap = cv2.VideoCapture('traffic.mp4')
ret, frame = cap.read()
# проверка успешности открытия видеофайла
if (cap.isOpened() == False): 
  print("Ошибка открытия файла")

# получение fps файла
fps = cap.get(cv2.CAP_PROP_FPS)
# print(fps)


fgbg = cv2.createBackgroundSubtractorMOG2()
l=0
r=0
while(cap.isOpened()):
  ret, frame = cap.read()

  if ret == True:     
    # корректировка скорости вывода
    time.sleep(1/fps)
   
    # отображение серого видео
    
    grayBlur = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # Перевод массива кадров в черно-белую градацию
      
    fgmask = fgbg.apply(grayBlur)   
    
    contours, hierarchy = cv2.findContours(fgmask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    cv2.line(frame, (390,490), (1200,440), (0,0,255), 5)
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area>1000:
            peri = cv2.arcLength(cnt,True)
            approx = cv2.approxPolyDP(cnt, 0.02*peri, True)
            x, y, w, h = cv2.boundingRect(approx)
            cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 5)
            if area>2000:
                M = cv2.moments (approx)
                if M["m00"]!=0:
                    approx_x= int(M["m10"]/M["m00"])
                    approx_y= int(M["m01"]/M["m00"])
                    #cv2.circle (frame, (approx_x, approx_y), 3, (0,255,0), 2)
                    if (approx_y==362) or (approx_y==363) or (approx_y==365) or (approx_y==366) or (approx_y==364) or (approx_y==367) or (approx_y==3):
                        if approx_x<660:
                            l=l+1
                        else:
                            r=r+1
    
               
    cv2.imshow("Objekt",frame) # обработанный вариант  
        
    # триггер для выхода
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break
 
  # выход из цикла
  else:
    break 
print("По правой полосе проехало:", r, "По левой полосе проехало:", l) 
# Освобожение памяти
cap.release()
# закрытие окна
cv2.destroyAllWindows() 
```
