```python
#Открытие видео
import cv2
import numpy as np
import time
print( cv2.__version__ )
# создание объекта типа VideoCapture
cap = cv2.VideoCapture('traffic.mp4')
# проверка успешности открытия видеофайла
if (cap.isOpened() == False): 
  print("Ошибка открытия файла")

# получение fps файла
fps = cap.get(cv2.CAP_PROP_FPS)
# print(fps)

while(cap.isOpened()):
  ret, frame = cap.read()

  if ret == True: 
        
    # корректировка скорости вывода
    time.sleep(1/fps)

    # отображение текущего кадра   
    cv2.imshow('frame',frame)  

    # триггер для выхода
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break
 
  # выход из цикла
  else:
    break 
 
# Освобожение памяти
cap.release()
 
# закрытие окна
cv2.destroyAllWindows() 
```

```python
#Добавление прямоугольгиков без обработки Canny, Blur и т.д.
import cv2
import numpy as np
import time
print( cv2.__version__ )
# создание объекта типа VideoCapture
cap = cv2.VideoCapture('traffic.mp4')
ret, frame = cap.read()
# проверка успешности открытия видеофайла
if (cap.isOpened() == False): 
  print("Ошибка открытия файла")

# получение fps файла
fps = cap.get(cv2.CAP_PROP_FPS)
# print(fps)


fgbg = cv2.createBackgroundSubtractorMOG2()

while(cap.isOpened()):
  ret, frame = cap.read()

  if ret == True:     
    # корректировка скорости вывода
    time.sleep(1/fps)
   
    # отображение серого видео
    
    grayBlur = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # Перевод массива кадров в черно-белую градацию
      
    fgmask = fgbg.apply(grayBlur)   
    
    contours, hierarchy = cv2.findContours(fgmask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area>1000:
            peri = cv2.arcLength(cnt,True)
            approx = cv2.approxPolyDP(cnt, 0.02*peri, True)
            x, y, w, h = cv2.boundingRect(approx)
            cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 5)
    
    
    cv2.imshow("Objekt",frame) # обработанный вариант  
        
    # триггер для выхода
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break
 
  # выход из цикла
  else:
    break 
 
# Освобожение памяти
cap.release()
# закрытие окна
cv2.destroyAllWindows() 
```

```python
#Добавление прямоугольгиков с обработкой
import cv2
import numpy as np
import time
print( cv2.__version__ )
# создание объекта типа VideoCapture
cap = cv2.VideoCapture('traffic.mp4')
ret, frame = cap.read()
# проверка успешности открытия видеофайла
if (cap.isOpened() == False): 
  print("Ошибка открытия файла")

# получение fps файла
fps = cap.get(cv2.CAP_PROP_FPS)
# print(fps)


fgbg = cv2.createBackgroundSubtractorMOG2()

while(cap.isOpened()):
  ret, frame = cap.read()

  if ret == True:     
    # корректировка скорости вывода
    time.sleep(1/fps)
   
    # отображение серого видео
    
    frameBlur = cv2.GaussianBlur(frame,(7,7),1)
    grayBlur = cv2.cvtColor(frameBlur, cv2.COLOR_BGR2GRAY) # Перевод массива кадров в черно-белую градацию
    
    threshold1 = 100
    threshold2 = 100
    
    frameCanny = cv2.Canny(grayBlur,threshold1, threshold2)
    
    kernel= np.ones((3,3))
    frameDil = cv2.dilate(frameCanny,kernel, iterations=1)
    
    
    
    fgmask = fgbg.apply(frameDil)
    
    
    
    contours, hierarchy = cv2.findContours(fgmask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area>1000:
            peri = cv2.arcLength(cnt,True)
            approx = cv2.approxPolyDP(cnt, 0.02*peri, True)
            x, y, w, h = cv2.boundingRect(approx)
            cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 5)
    
    
    cv2.imshow("Objekt",frame) # обработанный вариант
    
    
    
    
    # триггер для выхода
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break
 
  # выход из цикла
  else:
    break 
 
# Освобожение памяти
cap.release()
# закрытие окна
cv2.destroyAllWindows() 
```
