```python
import cv2
import numpy as np
import time


cap = cv2.VideoCapture('test.mp4')
obj_cascade=cv2.CascadeClassifier('haarcascade_fullbody.xml')
# проверка успешности открытия видеофайла


# получение fps файла
fps = cap.get(cv2.CAP_PROP_FPS)
# print(fps)
while True:
    ret,frame=cap.read()
    cap_gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    h,w = cap_gray.shape[:2]
    res=cv2.resize(frame, (int(0.5*w), int(0.5*h)), interpolation = cv2.INTER_CUBIC)
    res_gray=cv2.resize(cap_gray, (int(0.5*w), int(0.5*h)), interpolation = cv2.INTER_CUBIC)

    if frame is None:
        break
    if ret == True:
        # корректировка скорости вывода
        time.sleep(0.25/fps)
        
    #Создание маски нахождения объектов на проекзжей части    
    oblast=np.array([[0,155],[0,360],[450,130],[230,110]], np.int32)
    oblast=oblast.reshape(-1,1,2)
    mask=cv2.fillPoly(res.copy(), [oblast], (0,0,0))
    
    #Применение исключающего или
    xor=cv2.bitwise_xor(res,mask)
   
    #Перевод изображения в черно-белый формат
    res_bl = cv2.cvtColor(res, cv2.COLOR_BGR2GRAY)
    mask_bl = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
    
    #Поиск объектов на изображении
    objects=obj_cascade.detectMultiScale(mask_bl, scaleFactor=1.07, minNeighbors=5, minSize=(25,25),flags=0)
    for x,y,wigth, height in objects:
        cv2.rectangle(res, (x,y), (x+wigth,y+height), color=(0,255,0), thickness=2)
    #Поиск объектов на дороге
    objects_bad=obj_cascade.detectMultiScale(xor, scaleFactor=1.05, minNeighbors=1, minSize=(50,50),flags=0)
    for x_bad,y_bad,wigth_bad, height_bad in objects_bad:
        cv2.rectangle(res, (x_bad,y_bad), (x_bad+wigth_bad,y_bad+height_bad), color=(0,0,255), thickness=2)
        object_bad_color=res[y_bad: y_bad+height_bad, x_bad: x_bad+wigth_bad]

    
    #Отображение границ
    cv2.polylines(res, [oblast], True, (0,0,255))
    
    #Запрос времени
    if len(objects_bad)>0:
        realTime=time.asctime()
        realTime=time.strftime("%A %d %m %Y, %H. %M. %S")
    #Извлечение и сохранение кадра
        cv2.imwrite(str(realTime)+str(".jpg"), object_bad_color)
        
    # отображение текущего кадра 
    cv2.imshow('frame', res)  
    # триггер для выхода
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
# Освобожение памяти
cap.release()
 
# закрытие окна
cv2.destroyAllWindows() 
```
