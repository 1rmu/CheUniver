```python
# Работа по "Алгоритмы обработки изображения"
# Бинаризация
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('gradient.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width = img.shape[:2]
res = cv2.resize(img,(int(0.5*width), int(0.5*height)), interpolation = cv2.INTER_CUBIC)
gray = cv2.cvtColor(res, cv2.COLOR_BGR2GRAY)
ret,thresh1 = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(gray,127,255,cv2.THRESH_TOZERO)
ret,thresh4 = cv2.threshold(gray,127,255,cv2.THRESH_TOZERO_INV)
titles = ['Original Image','GrayScale','BINARY','BINARY_INV','TOZERO','TOZERO_INV']
images = [img_rgb, gray,thresh1, thresh2, thresh3, thresh4]
plt.figure(figsize=(20,10))

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
# Создание битовой маски 
# битовая маска может быть получена любым методом бинаризации, в котором 
# в результирующем изображении остаются всего два возможных значения пикселей - 0 либо 1
cv2.imshow("Image", thresh1)
cv2.waitKey(0)
# закрытие окна
cv2.destroyAllWindows()
```

```python
# сложение изображения с битовой маской
add_res = cv2.bitwise_or(res,cv2.cvtColor(thresh1, cv2.COLOR_GRAY2RGB))
cv2.imshow("Bitwise OR", add_res)
cv2.waitKey(0)# закрытие окна
cv2.destroyAllWindows() 
```

```python
# перемножение изображения с битовой маской
mul_res = cv2.bitwise_and(res,cv2.cvtColor(thresh1, cv2.COLOR_GRAY2RGB))
cv2.imshow("Bitwise AND", mul_res)
cv2.imshow("Bitwise AND1", res)
cv2.waitKey(0)# закрытие окна
cv2.destroyAllWindows() 
```

```python
# исключающее или изображения с битовой маской
xor_res = cv2.bitwise_xor(res,cv2.cvtColor(thresh1, cv2.COLOR_GRAY2RGB))
cv2.imshow("Bitwise XOR", xor_res)
cv2.waitKey(0)# закрытие окна
cv2.destroyAllWindows() 
```

```python
# морфологические операции с изображениями
# Эрозия
morph_img = cv2.imread('J.png',0)
height, width = morph_img.shape[:2]
morph_res = cv2.resize(morph_img,(int(3*width), int(3*height)), interpolation = cv2.INTER_CUBIC)
kernel = np.ones((3,3),np.uint8)
erosion = cv2.erode(morph_res,kernel,iterations = 1)
cv2.imshow("Original", morph_res)
cv2.imshow("Erosion", erosion)
cv2.waitKey(0)# закрытие окна
cv2.destroyAllWindows() 
```

```python
#Дилатация
dilation = cv2.dilate(morph_res,kernel,iterations = 1)
cv2.imshow("Original", morph_res)
cv2.imshow("Dilation", dilation)
cv2.waitKey(0)# закрытие окна
cv2.destroyAllWindows() 
```

```python
# Opening
morph_img = cv2.imread('J2.png',0)
height, width = morph_img.shape[:2]
morph_res = cv2.resize(morph_img,(int(3*width), int(3*height)), interpolation = cv2.INTER_CUBIC)
kernel = np.ones((10,10),np.uint8)
opening = cv2.morphologyEx(morph_res, cv2.MORPH_OPEN, kernel)
cv2.imshow("Original", morph_res)
cv2.imshow("Opening", opening)
cv2.waitKey(0)# закрытие окна
cv2.destroyAllWindows() 
```

```python
# Closing
morph_img = cv2.imread('J3.png',0)
height, width = morph_img.shape[:2]
morph_res = cv2.resize(morph_img,(int(3*width), int(3*height)), interpolation = cv2.INTER_CUBIC)
kernel = np.ones((10,10),np.uint8)
closing = cv2.morphologyEx(morph_res, cv2.MORPH_CLOSE, kernel)
cv2.imshow("Original", morph_res)
cv2.imshow("Closing", closing)
cv2.waitKey(0)# закрытие окна
cv2.destroyAllWindows() 
```
