```python
#Открытие исходного изображения

from PIL import Image
import imutils
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread("pr3.png")
cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
#Масштабирование изображения

image = cv2.imread("pr3.png")
height, width = image.shape[:2]
res = cv2.resize(image,(int(width), int(0.5*height)), interpolation = cv2.INTER_CUBIC)
cv2.imshow("Image", res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Перевод изображение в 8-ми битовую форму

image_gray = cv2.cvtColor(res, cv2.COLOR_BGR2GRAY)
cv2.imshow("Image_original", res)
cv2.imshow("Image_gray", image_gray)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
#Поиск горизонтальных линий и преобразование Хафа

blur = cv2.GaussianBlur(image_gray, (3,3),0)
thresh = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11,2)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
closed = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

Image_line=res
A=0
lines = cv2.HoughLinesP(closed,1,np.pi/180,100, minLineLength=100,  maxLineGap=20)
for line in lines:
    x1,y1,x2,y2 = line[0]
    if y1==y2:
        if y1%132==0:
            A=A+1
            cv2.line(Image_line, (x1,y1),(x2,y2),(0,0,255),3)
            
cv2.imshow("Image_line", Image_line)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```
